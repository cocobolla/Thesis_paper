# 분석을 위한 DataFrame Pickle 생성 스크립트

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
file_path = r'C:\Users\USER\Desktop\paper\kospi_10yr.csv'
df = pd.read_csv('./data/kospi_10yr.csv', encoding='cp949', thousands=',')
df
```

## 데이터 프레임 재구조화(Reshaping Data)

### 티커 - 주식 매칭 Series

```python
ticker_series = df.iloc[0,:]
ticker_series
```

```python
unique_ticker = ticker_series[1:].index.str.replace('\.[1-9]', '').unique()
ticker_series = ticker_series[unique_ticker]
ticker_series
```

### 관심있는 행과 열만 선택

```python
target_df = df.drop([0,1,2,4], axis=0)
target_df = target_df.reset_index(drop=True)
target_df
```

### Multi Index 재구조화

```python
target_col = ['Item Name', '수정주가(원)', '시가총액 (52주 평균)(백만원)']
target_df = target_df.loc[:, target_df.iloc[0,:].isin(target_col)]
target_df
```

```python
target_df.columns = target_df.columns.str.replace('\.[4,6]', '')
target_df.columns = pd.MultiIndex.from_arrays((target_df.iloc[0],target_df.columns), names=('Item', 'Symbol'))
target_df = target_df.drop([0], axis=0)
target_df.sort_index(axis=1, level=0, inplace=True)
target_df
```

```python
target_df[('Item Name', 'Symbol')] = target_df[('Item Name', 'Symbol')].astype('datetime64[ns]')
target_df = target_df.set_index(('Item Name', 'Symbol'))
target_df.index.rename('Date', inplace=True)
target_df
```

### 정재된 데이터프레임 분석

```python
# Thousands Point 제거
for col in target_df.columns:
    target_df[col] = target_df[col].astype('str').str.replace(",","").astype(float)  # 바로 .str하면 Nan 포함된 열에서 에러 발생
target_df
```
